Hmmmm this is a bit odd. Why are laname20 values missing? The values in column 25 are obviously laname20 values, but it's odd they're missing. Would expect this in the 2020_geog21 csv file. Let's load up that file and take a look.

```python
df21 = pd.read_csv('../data/external/MYEB1_detailed_population_estimates_series_UK_(2020_geog21).csv')
df21.info()
```

#### First thing to notice is that there isn't a column 25.

#### I've tracked down some newer PFA area data from the ONS which uses 2021 LA codes, so let's stick with this dataset rather than the one which uses 2020 LA codes.

#### Filtering the dataset to only include England and Wales

```python
df21
```

```python
filt = df21['country'].str.contains("(?:^E|^W)", regex=True)
df_eng_wales= df21[filt]
df_eng_wales
```

#### Further filtering to only include adult women

```python
filt1 = df_eng_wales['age'] >= 18
filt2 = df_eng_wales['sex'] == 2
filt = filt1 & filt2
df3 = df_eng_wales[filt]
df3
```

#### Melting the columns to make the data long, rather than wide

```python
df4 = df3.melt(id_vars=["ladcode21", "laname21", "country", "sex", "age"], var_name="year", value_name="population")
df4
```

#### Removing 'population_' from the `year` values

```python
df4['year'] = df4['year'].str.replace("population_","", regex=True)
```

```python
df4
```

#### Aggregating all of the ages to just have one population figure for adult women in each local authority for each year

```python
df5 = df4.groupby(['ladcode21', 'laname21', 'year'], as_index=False, sort=False).agg({'population':'sum'})
df5
```

#### Converting `year` values to integers to allow filtering

```python
df5['year']= df5['year'].astype('int')
df5.dtypes
```

#### Just filtering some values to check against values in my previous workbook.

```python
filt1 = df5['year'] >= 2017
filt2 = df5['laname21'] == 'Blackpool'
filt = filt1 & filt2
df5[filt]
```

#### Let's also check some of those new unitary LAs and see what values we have.

```python
filt = df5['laname21'] == 'North Northamptonshire'
df5[filt]
```

```python
filt = df5['laname21'] == 'West Northamptonshire'
df5[filt]
```

#### We have values for the entire time series which is good, but further assurance will need to be conducted following the merging with PFA areas.

#### Let's now save this out to a new csv file.

```python
df5.to_csv('../data/interim/LA_population_female_2001_2020-cleansed.csv', index=False)
```

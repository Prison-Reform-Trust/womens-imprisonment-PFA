# Women's imprisonment rates
## Criminal Justice Statistics Police Force Area: Custodial, community and suspended sentences

#### Importing pandas library and reading in data

```python
import pandas as pd
df = pd.read_csv('../data/interim/PFA_2009-21_women_cust_comm_sus.csv')
```

```python
df.head()
```

```python
df.drop(columns="sentence_len", inplace=True)
df.head()
```

### Grouping

```python
df2 = df.groupby(['pfa', 'year', 'outcome'], as_index=False)['freq'].sum()
df2
```

#### Checking data before 2020

```python
df2.query('pfa == "Derbyshire" & outcome == "Immediate custody"')
```

```python
df2.query('pfa == "Cambridgeshire" & outcome == "Suspended sentence"')
```

#### Outputting to CSV

```python
df2.to_csv('../data/processed/PFA_2009-21_women_cust_comm_sus_FINAL.csv', index=False)
```
